```python
num_to_class={  
  "1"  : "N",
  "2"  : "A",   
  "3"  : "NUM",   
  "4"  : "V", 
  "5"  : "ART",   
  "6"  : "PRON",   
  "7"  : "ADV",  
  "8"  : "PREP",   
  "9"  : "C",
  "10" : "I",
  "11" : "SCON",
  "12" : "CCON",
  "13" : "LET",
  "14" : "ABB",   
  "15" : "TO"
}
```

```python
wfdict=defaultdict(dict)

celex_path = os.path.join("data", "Celex", "Celex", "english")

with open(os.path.join(celex_path, "eow", "eow.cd"),"r") as f_eow:
    for line in f_eow.readlines():
        fields=line.strip().split("\\")
        wf_idnum=fields[0]
        wf_idnumlemma=fields[3]
        wf_worddia=fields[1]
        wfdict[wf_idnum]["worddia"]=wf_worddia
        wfdict[wf_idnum]["idnumlemma"]=wf_idnumlemma 

lemmadict=defaultdict(dict)

with open(os.path.join(celex_path, "esl", "esl.cd"),"r") as f_esl:
    for line in f_esl.readlines():
        fields=line.strip().split("\\")
        lemma_idnum=fields[0]
        lemma_head=fields[1]
        lemma_classnum=fields[3]
        lemmadict[lemma_idnum]["head"]=lemma_head
        lemmadict[lemma_idnum]["class"]=num_to_class[lemma_classnum]

with open(os.path.join(celex_path, "eml", "eml.cd"),"r") as f_eml:
    for line in f_eml.readlines():
        fields=line.strip().split("\\")
        lemma_idnum=fields[0]
        lemma_morphstatus=fields[3]
        lemmadict[lemma_idnum]["morphstatus"]=lemma_morphstatus

with open(os.path.join(celex_path, "eol", "eol.cd"),"r") as f_eol:
    for line in f_eol.readlines():
        fields=line.strip().split("\\")
        lemma_idnum=fields[0]
        syllables=fields[7]
        lemmadict[lemma_idnum]["syllables"]=syllables

with open(os.path.join(celex_path, "epl", "epl.cd"), "r") as f_epl:
    for line in f_epl.readlines():
        fields=line.strip().split("\\")
        lemma_idnum=fields[0]
        phontrans=fields[5]
        lemmadict[lemma_idnum]["phontrans"]=phontrans
```

```python
for wf_idnum, wf_info in wfdict.items():
    idnumlemma=wf_info["idnumlemma"]
    wf_info["lemma"]=lemmadict[idnumlemma]["head"]
    wf_info["class"]=lemmadict[idnumlemma]["class"]
    wf_info["morphstatus"]=lemmadict[idnumlemma]["morphstatus"]
    wf_info["syllables"]=lemmadict[idnumlemma]["syllables"]
    wf_info["phontrans"]=lemmadict[idnumlemma]["phontrans"]
```

```python
monomorph_is_lemma_list = []
not_m_is_l = []

for word in wfdict:
    if wfdict[word]['morphstatus'] == 'M' and wfdict[word]['worddia'] == wfdict[word]['lemma']:
        monomorph_is_lemma_list.append(wfdict[word]['worddia'])
    else:
        not_m_is_l.append(wfdict[word]['worddia'])

print(f"{len(monomorph_is_lemma_list)} are monomorphemic lemmas:\n{monomorph_is_lemma_list}")
print(f"{len(not_m_is_l)} are NOT monomorphemic lemmas:\n{not_m_is_l[:10]}")
```
